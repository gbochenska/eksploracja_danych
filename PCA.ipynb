# Stroke

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/stroke_data.csv")
none_counts = data.isna().sum()

print("Liczba wartości None w poszczególnych kolumnach:")
print(none_counts)
```

```python
data_cleaned = data.dropna()

print("Liczba wartości None w poszczególnych kolumnach po usunięciu wierszy:")
print(data_cleaned.isna().sum())
```

```python
X = data_cleaned.iloc[:, :-1]
y = data_cleaned.iloc[:, -1]

# Standaryzacja danych
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Stosunek wariancji')
plt.xlabel('Główne składowe')
plt.ylabel('Wariancja')
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=8)
X_pca = pca.fit_transform(X_scaled)
```
